# Tokenizing text into unigrams & bigrams using Gensim

```python
from gensim.models import Phrases
from gensim.models.phrases import Phraser
from gensim import models
# Tokenize(split) the sentences into words
data = df[df['word_count']>= 5]
#data = data.iloc[0:100]
data = data.reset_index(drop=True)
data['index']= data.index
documents = data['clean_text'].astype(str).to_list()
tokenized_list = [[text for text in doc.split()] for doc in documents]

bigram = Phrases(tokenized_list, min_count=1, threshold=2,delimiter=b' ')
bigram_phraser = Phraser(bigram)

bigram_token = []
for sent in tokenized_list:
    bigram_token.append(bigram_phraser[sent])


# Create the Dictionary and Corpus with unigram
#bow = tokenized_list
#mydict = corpora.Dictionary(bow)
#corpus = [mydict.doc2bow(doc) for doc in bow]

# Create the Dictionary and Corpus including both unigram & bigrams
bow = bigram_token
mydict = gensim.corpora.Dictionary(bow)
corpus = [mydict.doc2bow(doc) for doc in bow]
```

# Build LDA Model & Select Number of Topics based on Coherence Score

```python
from gensim.models import LdaModel
def compute_coherence_values_symmetric(dictionary, corpus, texts, alpha_type,  limit, start , step):
    
    coherence_values = []
    model_list = []
    for num_topics in range(start, limit, step):
        #model = gensim.models.wrappers.LdaMallet(mallet_path, corpus=corpus, num_topics=num_topics, id2word=id2word)
        model = LdaModel(corpus=corpus, id2word=mydict, random_state=100,num_topics=num_topics,
                          passes=5,chunksize=1000, alpha=alpha_type, decay=0.5, offset=64, eta='auto',
                          eval_every=0, iterations=100, gamma_threshold=0.001, per_word_topics=True )

        model_list.append(model)
        coherencemodel = CoherenceModel(model=model, texts=texts, dictionary=dictionary, coherence='c_v')
        coherence_values.append(coherencemodel.get_coherence())

    return model_list, coherence_values
```

```python
# Can take a long time to run.
limit=45; start=5; step=5; alpha_type = 'auto'  #'asymmetric', 'auto'
model_list, coherence_values = compute_coherence_values_symmetric(dictionary=mydict, corpus=corpus,
                                                        texts=bow, alpha_type = 'auto',
                                                        start= start, limit=limit, step=step)

# Show graph
x = range(start, limit, step)
plt.plot(x, coherence_values)
plt.xlabel("Num Topics")
plt.ylabel("Coherence score")
plt.legend(("coherence_values"), loc='best')
plt.show()
```

```python
# Print the coherence scores
for m, cv in zip(x, coherence_values):
    print("Num Topics =", m, " has Coherence Value of", round(cv, 4))
```

### 25 topics give reasonable high coherence score

```python
lda_model = LdaModel(corpus=corpus, id2word=mydict, random_state=100,num_topics=25,
                 passes=5,chunksize=1000, alpha='auto', decay=0.5, offset=64, eta='auto',
                 eval_every=0, iterations=100, gamma_threshold=0.001, per_word_topics=True)
```

```python
coherencemodel = CoherenceModel(model=lda_model, texts=bow, dictionary=mydict, coherence='c_v')
print('coherence score is: ',coherencemodel.get_coherence())
```

# Get Sentence Level Topic Distribution

```python
def format_topics_sentences(lda_model=lda_model, corpus=corpus, texts=documents):
    # Init output
    sent_topics_df = pd.DataFrame()

    # Get main topic in each document
    for i, row in enumerate(lda_model[corpus]):
        row = sorted(row[0], key=lambda x: (x[1]), reverse=True)
        # Get the Dominant topic, Perc Contribution and Keywords for each document
        for j, (topic_num, prop_topic) in enumerate(row):
            if j == 0:  # => dominant topic
                wp = lda_model.show_topic(topic_num)
                topic_keywords = ", ".join([word for word, prop in wp])
                sent_topics_df = sent_topics_df.append(pd.Series([int(topic_num), round(prop_topic,4), topic_keywords]), ignore_index=True)
            else:
                break
    sent_topics_df.columns = ['Dominant_Topic', 'Perc_Contribution', 'Topic_Keywords']

    # Add original text to the end of the output
    #contents = pd.Series(texts)
    #sent_topics_df = pd.concat([sent_topics_df, contents], axis=1)
    return(sent_topics_df)


df_topic_sents_keywords = format_topics_sentences(lda_model=lda_model, corpus=corpus, texts=documents)

# Format
df_dominant_topic = df_topic_sents_keywords.reset_index()
df_dominant_topic.columns = ['index', 'Dominant_Topic', 'Topic_Perc_Contrib', 'Keywords']
```

```python
df_topic = pd.merge(data, df_dominant_topic, on = 'index')
df_topic.head()[['index', 'Dominant_Topic', 'Topic_Perc_Contrib', 'Keywords', 'text']]
```

```python
def topic_distribution_sentence(lda_model=lda_model, corpus = corpus):
    get_document_topics = lda_model.get_document_topics(corpus)
    topic_distribution_df = pd.DataFrame()
    for i, row in enumerate(get_document_topics):
        items ={}
        row_df = []
        for (topic_num, perc) in row:
            items[topic_num]= round(perc,4)
        row_df = pd.DataFrame([items])
        row_df.columns = ['percentage_topic_' + str(col) for col in row_df.columns]
        row_df['index']=i
        topic_distribution_df = pd.concat([row_df,topic_distribution_df])
    topic_distribution_df = topic_distribution_df.fillna(0)
    return topic_distribution_df
```

```python
#get_document_topics = lda_model.get_document_topics(corpus[0:10])
#print(get_document_topics[0])
topic_distribution_df = topic_distribution_sentence(lda_model=lda_model, corpus = corpus)
```

```python
df_topic_vect=pd.merge(df_topic, topic_distribution_df,  on = 'index')
df_topic_vect.to_csv('LDA_25_topic_distribution.csv')
```
