```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'svg'
plt.style.use('ggplot')
%matplotlib inline
```

```python
df = pd.read_csv("Data/RetailCustomerSales.csv", sep=';')
df.head()
```

```python
df.isnull().sum()
```

```python
df[["CustomerID", "Profession", "ItemCategory1", "ItemCategory2", 
    "ItemCategory3", "HaveChildren"]] = df[["CustomerID", "Profession", "ItemCategory1", "ItemCategory2",
                            "ItemCategory3", "HaveChildren"]].astype(str) 
df.describe(include =[object])
```

```python
df['Amount'].hist()
```

```python
df_customer = df[["CustomerID","Sex","Age","Profession","CityType","YearsInCity","HaveChildren"]].drop_duplicates()
df_customer.head()
```

```python
fig, axes = plt.subplots(2, 3, figsize=(12, 4))
features_customer = df_customer.columns[1:]
for feature, ax in zip(features_customer, axes.ravel()):
    df_customer[feature].value_counts().plot.bar(ax=ax)
    ax.set_title(feature)
plt.tight_layout()
```

```python
df_customer = df_customer.sort_values(by ='CityType')
sns.catplot(y = "Age", col = "CityType", kind = "count", data = df_customer
           ,order = df_customer["Age"].value_counts().index, sharex=False)
```

```python
df_agg_1 = df.groupby(['CustomerID','ItemCategory1']).agg({'Amount':['sum'],
                                                'ItemID':['count']
    
}).reset_index()
#df_agg.columns = df_agg.columns.droplevel(1)
df_agg_1.columns = ['CustomerID','ItemCategory1','Item_amount','Item_count']
df_agg_1['Item_amount_mean'] = df_agg_1['Item_amount']/df_agg_1['Item_count']
df_agg_1.head()
```

```python
df_1 = pd.pivot_table(df_agg_1, index = 'CustomerID', columns = 'ItemCategory1'
               ,values =[ 'Item_count','Item_amount'],fill_value=0)
df_1.columns = ['_'.join(col).strip() for col in df_1.columns.values]
df_1 = df_1.reset_index()
df_1.head()
```

```python
df_1.columns
```

```python
features_count = [
    'Item_count_1', 'Item_count_10', 'Item_count_11',
       'Item_count_12', 'Item_count_13', 'Item_count_14', 'Item_count_15',
       'Item_count_16', 'Item_count_17', 'Item_count_18', 'Item_count_2',
       'Item_count_3', 'Item_count_4', 'Item_count_5', 'Item_count_6',
       'Item_count_7', 'Item_count_8', 'Item_count_9'
]
df_1[features_count].describe()
```

```python
corr = df_1[features_count].corr()
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True
with sns.axes_style("white"):
    f, ax = plt.subplots(figsize=(10, 8))
    sns.heatmap(corr, annot=True, cmap = 'coolwarm', mask=mask)
```
