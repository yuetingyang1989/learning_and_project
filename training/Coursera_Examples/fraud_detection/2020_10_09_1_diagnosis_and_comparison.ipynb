```python
EstimatesType = types.StructType([
    types.StructField("point_est", types.DoubleType()),
    types.StructField("conf_int_lb", types.DoubleType()),
    types.StructField("conf_int_ub", types.DoubleType()),
])

def estimates(rows, point_estimator):
    data = pd.DataFrame(row.asDict() for row in rows)
    result = project_utils.get_point_estimate_and_confidence_interval(
        point_estimator, data
    )
    return float(result.point_est), float(result.conf_int_lb), float(result.conf_int_ub)

def fraud_rate(data: pd.DataFrame) -> float:
    return float(np.average(data["binary_label"]))

fraud_rate_estimates_udf = sf.udf(
    f=functools.partial(estimates, point_estimator=fraud_rate), 
    returnType=EstimatesType
)
```

```python
fraud_rate = (
    weekly_aggs_df
    .withColumn("fraud_rate", fraud_rate_estimates_udf(sf.col("rows")))
    .select("week_start_date", "fraud_rate.*")
    .withColumnRenamed("point_est", "fraud_rate")
    .withColumnRenamed("conf_int_lb", "ci_90_lb")
    .withColumnRenamed("conf_int_ub", "ci_90_ub")
    .sort("week_start_date")
    .toPandas()
)
```

```python
def plot_series_with_confint(pdf, y=None, color="blue", title=None, as_percentage=False, 
                             mark_periods: Optional[List[Tuple[datetime.datetime, datetime.datetime]]]=None, 
                             mark_periods_color: str="red", mark_label=None):
    y = pdf.columns[1]
    factor = 100 if as_percentage else 1
    pe = factor*pdf[y] # point estimate
    ub = factor*pdf["ci_90_ub"] # 90% CL upper bound
    lb = factor*pdf["ci_90_lb"] # 90% CL lower bound
    date = pdf["week_start_date"]
 
    # For visual purposes, first plot marked periods
    if mark_periods:
        for start_date, end_date in mark_periods:
            plt.axvspan(
                xmin=start_date,
                xmax=end_date,
                color=mark_periods_color,
                alpha=0.2,
                label=mark_label
            )
            mark_label = None # so that it only appears once

    plt.plot(date, ub, color=color, alpha=0.1)
    plt.plot(date, lb, color=color, alpha=0.1)
    plt.plot(date, pe, color=color, label=y, marker="o", markersize=3)

    plt.fill_between(
                x=date,
                y1=lb,
                y2=ub,
                alpha=0.2,
                color=color
            )
    if title is None:
        title = f"{y!r} with 90% CL interval"
    plt.title(title)
    plt.legend()
    plt.xticks(rotation=90)
    plt.xlabel("Date")
    ylabel = f"{y} (%)" if as_percentage else y
    plt.ylabel(ylabel)
    plt.grid()
```

```python
counts = (
    weekly_aggs_df
    .select("week_start_date", "non_fraud_count", "fraud_count")
    .sort("week_start_date")
    .toPandas()
)
```

```python
NUM_TRAINING_WEEKS = 9*4
NUM_WEEKS_BETWEEN_TRAIN_AND_TEST = 3*4
TEST_START_INDEX = NUM_TRAINING_WEEKS + NUM_WEEKS_BETWEEN_TRAIN_AND_TEST
chi2_columns = [
    "test_week_start_date", 
    "test_non_fraud_count", 
    "test_fraud_count",
    "train_non_fraud_count",
    "train_fraud_count",
    "chi2",
    "p"
]
Chi2Row = namedtuple("Chi2Row", chi2_columns)
chi2_rows = []
for train_start_index, (_, test_row) in enumerate(counts[TEST_START_INDEX:].iterrows()):
    test_week_start_date, test_non_fraud_count, test_fraud_count = test_row
    
    train_counts = counts[train_start_index: train_start_index + NUM_TRAINING_WEEKS]
    train_non_fraud_count = np.sum(train_counts["non_fraud_count"])
    train_fraud_count = np.sum(train_counts["fraud_count"])
    
    chi2, p, _, _ = scipy.stats.chi2_contingency([
        [train_non_fraud_count, train_fraud_count], 
        [test_non_fraud_count, test_fraud_count]
    ])

    chi2_rows.append(
        Chi2Row(
            test_week_start_date, 
            test_non_fraud_count, 
            test_fraud_count,
            train_non_fraud_count,
            train_fraud_count,
            chi2,
            p
        )
    )
```

```python
prevalence_chi2 = pd.DataFrame.from_records(chi2_rows, columns=chi2_columns)
```

```python
def plot_prevalance_chi2_p(prevalance_chi2):
    date = prevalence_chi2["test_week_start_date"]
    p = prevalence_chi2["p"]
    plt.plot(date, p, label="p-value")
    plt.xticks(rotation=90)
    plt.title("p-value for Pearson Chi2 test for train/test prevalence")
    plt.xlabel("Date")
    plt.ylabel("p-value")
    plt.axhline(0.05, color="black", linestyle="dashed")
    plt.grid()
```

```python
plot_prevalance_chi2_p(prevalence_chi2)
```

```python
def get_prior_shift_periods(prevalence_chi2: pd.DataFrame, significance_level=0.05):
    periods = []
    for (_, row), (_, next_row) in zip(prevalence_chi2[:-1].iterrows(), prevalence_chi2[1:].iterrows()):
        if (row["p"] > significance_level) and (next_row["p"] <= significance_level):
            start_date = next_row["test_week_start_date"]
        elif (row["p"] <= significance_level) and (next_row["p"] > significance_level):
            end_date = row["test_week_start_date"]
            periods.append((start_date, end_date))
        else:
            pass
    return periods
```

```python
first_test_week = prevalence_chi2.iloc[0]["test_week_start_date"]
date_selector = fraud_rate["week_start_date"] >= first_test_week
plot_series_with_confint(
    pdf=fraud_rate[date_selector],
    color="blue",
    as_percentage=True,
    mark_periods=get_prior_shift_periods(prevalence_chi2),
    mark_label="p(y|train)!=p(y|test)"
)
```

```python
# NUM_TRAINING_WEEKS = 9*4
# NUM_WEEKS_BETWEEN_TRAIN_AND_TEST = 3*4
# TEST_START_INDEX = NUM_TRAINING_WEEKS + NUM_WEEKS_BETWEEN_TRAIN_AND_TEST
TrainTestPeriods = namedtuple("TrainTestPeriods", "train_start_date, train_end_date, test_start_date, test_end_date")
def get_train_and_test_periods_for_test_week(test_week_start_date: datetime.datetime) -> TrainTestPeriods:
    assert test_week_start_date.isoweekday() == 1 # Monday
    train_start_delta = datetime.timedelta(weeks=NUM_TRAINING_WEEKS+NUM_WEEKS_BETWEEN_TRAIN_AND_TEST)
    train_end_delta = datetime.timedelta(weeks=NUM_WEEKS_BETWEEN_TRAIN_AND_TEST)
    return TrainTestPeriods(
        train_start_date = test_week_start_date - train_start_delta,
        train_end_date = test_week_start_date - train_end_delta,
        test_start_date = test_week_start_date,
        test_end_date = test_week_start_date + datetime.timedelta(days=6)
    )
```

```python
def get_week_selector_col(test_week_start_date: datetime.datetime) -> Column:
    train_test = get_train_and_test_periods_for_test_week(test_week_start_date)
    col_name = get_week_selector_col_name(test_week_start_date)
    return (
        sf
        .when(sf.col("yyyy_mm_dd").between(train_test.train_start_date, train_test.train_end_date), 1)
        .when(sf.col("yyyy_mm_dd").between(train_test.test_start_date, train_test.test_end_date), 1)
        .otherwise(0)
        .alias(col_name)
    )


def get_week_selector_col_name(test_week_start_date: datetime.datetime):
    yyyy_mm_dd = test_week_start_date.isoformat().replace("-", "_")
    return f"{yyyy_mm_dd}_selector"


def get_week_train_test_col(test_week_start_date: datetime.datetime) -> Column:
    train_test = get_train_and_test_periods_for_test_week(test_week_start_date)
    col_name = get_week_train_test_col_name(test_week_start_date)
    return (
        sf
        .when(sf.col("yyyy_mm_dd").between(train_test.train_start_date, train_test.train_end_date), 0)
        .when(sf.col("yyyy_mm_dd").between(train_test.test_start_date, train_test.test_end_date), 1)
        .otherwise(None)
        .alias(col_name)
    )


def get_week_train_test_col_name(test_week_start_date: datetime.datetime):
    yyyy_mm_dd = test_week_start_date.isoformat().replace("-", "_")
    return f"{yyyy_mm_dd}_train_test_label"
```

```python
label_cols = [get_week_train_test_col(test_week_start_date) for test_week_start_date in test_week_start_dates]
selector_cols = [get_week_selector_col(test_week_start_date) for test_week_start_date in test_week_start_dates]
```

```python
data_h2o_df = project_utils.spark_to_h2o_frame(data_df.select("*", *label_cols, *selector_cols), metadata)
```

```python
for test_week_start_date in test_week_start_dates:
    week_train_test_col_name = get_week_train_test_col_name(test_week_start_date)
    data_h2o_df[week_train_test_col_name] = data_h2o_df[week_train_test_col_name].asfactor()
    week_selector_col_name = get_week_selector_col_name(test_week_start_date)
    data_h2o_df[week_selector_col_name] = data_h2o_df[week_selector_col_name].asfactor()
```

```python
from h2o.estimators import H2ORandomForestEstimator

```

```python
models = {}
for test_week_start_date in test_week_start_dates[::4]:
    model = H2ORandomForestEstimator(nfolds=5)
    week_selector_col_name = get_week_selector_col_name(test_week_start_date)
    week_train_test_col_name = get_week_train_test_col_name(test_week_start_date)
    train_h2o_df = data_h2o_df[data_h2o_df[week_selector_col_name]]
    features = metadata["numerical_features"] + metadata["categorical_features"]
    model.train(x=features, y=week_train_test_col_name, training_frame=train_h2o_df)
    models[test_week_start_date] = model
```

```python
for test_week_start_date, model in models.items():
    print(f"{test_week_start_date}: {model.aucpr()} {model.varimp()[0]}")
```
