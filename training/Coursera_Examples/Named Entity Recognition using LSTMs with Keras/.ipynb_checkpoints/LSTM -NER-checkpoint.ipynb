### Task 1: Project Overview and Import Modules

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
np.random.seed(0)
plt.style.use("ggplot")

import tensorflow as tf
print('Tensorflow version:', tf.__version__)
print('GPU detected:', tf.config.list_physical_devices('GPU'))
```

### Task 2: Load and Explore the NER Dataset

*Essential info about tagged entities*:
- geo = Geographical Entity
- org = Organization
- per = Person
- gpe = Geopolitical Entity
- tim = Time indicator
- art = Artifact
- eve = Event
- nat = Natural Phenomenon

```python
data = pd.read_csv('ner_dataset.csv', encoding = 'latin1')
data = data.fillna(method = 'ffill') # fill the cell with the last missing value
data.head(20)
```

```python
print("Unique words in corpus:", data['Word'].nunique())
print("Unique tags in corpus:", data['Tag'].nunique())
```

```python
words = list(set(data['Word'].values))
words.append("ENDPAD")
num_words = len(words)
```

```python
tags = list(set(data["Tag"].values))
num_tags = len(tags)
```

```python
num_words, num_tags
```

### Task 3: Retrieve Sentences and Corresponsing Tags

```python
class SentenceGetter(object):
    def __init__(self, data):
        self.n_sent = 1
        self.data = data
        agg_func = lambda s: [(w,p,t) for w, p, t in zip(s["Word"].values.tolist(),
                                                        s["POS"].values.tolist(),
                                                        s["Tag"].values.tolist())]
        self.grouped = self.data.groupby("Sentence #").apply(agg_func)
        self.sentences = [s for s in self.grouped]
```

```python
getter = SentenceGetter(data)
sentences = getter.sentences
```

```python
sentences[0]
```

### Task 4: Define Mappings between Sentences and Tags

```python
word2idx = {w: i+1 for i, w in enumerate(words)}
tag2idx = {t: i for i, t in enumerate(tags)}
```

```python
word2idx
```

### Task 5: Padding Input Sentences and Creating Train/Test Splits

```python
# to choose the padding length, looking at the distribution of words per sentences
plt.hist([len(s) for s in sentences], bins = 50)
plt.show()
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import to_categorical

max_len = 50

# X is a numerical representation of our words
# 2 loops: 1 loop over sentences, 1 loop over words in each sentence
X = [[word2idx[w[0]] for w in s] for s in sentences]
X = pad_sequences(maxlen = max_len, sequences = X, 
                  padding = 'post', value = num_words -1 )

# y targat vector, 2 loops
y = [[tag2idx[w[2]] for w in s] for s in sentences]
y = pad_sequences(maxlen = max_len, sequences = y,
                 padding = 'post', value = tag2idx["O"])
y = [to_categorical(i, num_classes = num_tags) for i in y]
```

```python
from sklearn.model_selection import train_test_split

# validation set would be created later
x_train, x_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size = 0.1,
                                                   random_state = 1)
```

### Task 6: Build and Compile a Bidirectional LSTM Model

```python
from tensorflow.keras import Model, Input
from tensorflow.keras.layers import LSTM, Embedding, Dense
from tensorflow.keras.layers import TimeDistributed, SpatialDropout1D, Bidirectional
```

```python
input_word = Input(shape=(max_len, ))
# Create word embeddings
model = Embedding(input_dim = num_words, output_dim = max_len, 
                 input_length = max_len)(input_word)  #apply embedding on the input_word layer
# Create dropout layer
## normally dropout is applied on a layer in neural network and
## drop out randomly of values from previous layer, 
## the dropouts are independent of each other
## Spatial dropout layer is special, 
## e.g. if it drops the 2nd value in the input, 
## then it drops the 2nd value from all the channels in that layer
## instead of dropping a node, it drops the 1D features in all channels

model = SpatialDropout1D(0.1)(model) #apply dropout on previous defined model

model = Bidirectional(LSTM(units = 100, return_sequences= True,
                          recurrent_dropout= 0.1))(model)

#apply Dense layer on each unit, so 100 unit = 100 Dense application
out = TimeDistributed(Dense(num_tags, activation= 'softmax'))(model)

model = Model(input_word, out)

model.summary()
```

```python
model.compile(optimizer = 'adam', 
              loss = 'categorical_crossentropy',
             metrics = ['accuracy'])
```

### Task 7: Train the Model

```python
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
from livelossplot.tf_keras import PlotLossesCallback
```

```python
# list memory in GPU
# 11441MiB = 11GB
!nvidia-smi
```

```python
# Define early stopping callback
early_stopping = EarlyStopping(monitor = 'val_accuracy', 
                               patience = 1,# patience = 1 meaning if accuracy is not improving in the next 1 epoch, then stop
                               verbose = 0, # no output
                               mode = 'max', # we want to maximize accuracy
                               restore_best_weights = False
                              ) 
# Define all the callbacks and put them in a list
callbacks = [PlotLossesCallback(), early_stopping]

history = model.fit(
    x_train, np.array(y_train), #y_train is a list, has to be converted to np.array
    validation_split = 0.2, #further split training set to train & validation
    batch_size = 32, #if we have bigger GPU, we can set batch_size = 64
    epochs = 3,# you can set it to 100 and use the early stopping to callback and without hardcoding epoch number
    verbose = 1,
    callbacks = callbacks
)
```

### Task 8: Evaluate Named Entity Recognition Model

```python
model.evaluate(x_test, np.array(y_test))
```
