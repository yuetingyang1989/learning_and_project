##  product association 
using two different mechanisms (we aren't using lift here because of how similar the popularity levels are).

4. Top movies for someone who has seen Toy Story: Calculate movies that most often occur with Movie #1:  Toy Story, using the (x+y)/x method described in class. In other words, for each movie, calculate the percentage of Toy Story raters who also rated that movie. Order with the highest percentage first, and submit the top 3 (along with the correlations for the top two).  Note, you will have ties -- to break the ties, use the lowest-numbered movie as the higher-ranked one.  In other words, if Movies 541 and 318 are tied, then 318 gets the higher rank

```python
count_associated_movie = (df_ratings[df_ratings['1: Toy Story (1995)']>=0]>=0).sum().sort_values(ascending = False).to_frame(name = 'associated_count')
count_associated_movie[0:8]
```

```python
total_movie_counts = df_ratings.count().sort_values(ascending = False).to_frame(name = 'total_count')
total_movie_counts[0:4]
```

```python
df_toy = count_associated_movie.join( total_movie_counts)
df_toy['associated_perc'] = df_toy['associated_count']/df_toy['total_count']
df_toy = df_toy.sort_values(by = 'associated_perc', ascending = False)
df_toy.head(5)
```

## demographic analysis 

6. Mean rating difference by gender:  First, recompute the mean rating for each movie separately for males and for females.  And calculate the overall mean rating (across all ratings) for males and females. Submit the two movies that have the greatest differences (one where men are most above women, and one where women are most above men) along with the differences in average.  Also submit the difference in overall rating averages (female average - male average).

```python
df_gender = data[data.columns[~data.columns.isin(['User'])]].groupby(["Gender (1 =F, 0=M)"]).mean().T
df_gender['diff'] = df_gender[1] - df_gender[0] #female - male
df_gender = df_gender.sort_values(by = 'diff', ascending = False)
df_gender
```

```python
df_gender.describe()
```

7. Next, compute the % of ratings 4+ separately for males and females.  You'll be asked to submit two movies as above (largest difference in each direction). And again you'll indicate whether men or women are more likely to rate movies 4 stars or above.

```python
df_rating_gender = pd.DataFrame()
for gender in (0, 1):
    df_gender = data[data["Gender (1 =F, 0=M)"]== gender]
    df_rating = df_gender[df_gender.columns[~df_gender.columns.isin(['Gender (1 =F, 0=M)', 'User'])]]
    df_rating_gender_temp = ((df_rating >= 4).sum()/(df_rating >= 0).sum()).sort_values(ascending = False).to_frame(name = "rating")
    df_rating_gender_temp.reset_index(level=0, inplace=True)
    df_rating_gender_temp['gender'] = gender
#    print(df_rating_gender_temp.head())
    df_rating_gender = pd.concat([df_rating_gender,df_rating_gender_temp])
```

```python
df_rating_gender.head()
```

```python
df = df_rating_gender.pivot(index='index', columns='gender', values='rating').reset_index()
df['diff'] = df[1]- df[0]
df = df.sort_values(by = 'diff', ascending = False)
df.head()
```
